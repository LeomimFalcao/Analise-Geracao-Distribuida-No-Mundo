# **Análises Gráficas**

Neste tópico iremos observar como se comportam gráficamente certas situações: Consumo total de energia elétrica no mundo todo, País com menor consumo, País com maior consumo e o contexto do nosso país (Brasil). E a partir do exposto contextualizar e abordar pontos que são pertinentes.

```
plot.figure(figsize=(20,7))
plot.plot(dados['World'], c ='m')
plot.xlabel('Year')
plot.ylabel('Energy Consumption in MWh')
plot.show()
```

# Valores Referentes ao Consumo de Energia Elétrica no Brasil 
(Nosso contexto)

```
plot.figure(figsize=(12,6))
plot.plot(dados['Brazil'],c='y')
plot.xlabel('Year')
plot.ylabel('TWh/Year Consumption in Brazil')
plot.legend(labels=['Brazil'])
plot.title("Valores Referentes ao Consumo de Energia Elétrica no Brasil")
plot.show()
```

# Valores referentes ao Consumo de Energia Elétrica no Niue 
(País com menor consumo de energia elétrica dentro do estudo disponível)

```
plot.figure(figsize=(12,6))
plot.plot(dados['Niue'])
plot.xlabel('Year')
plot.ylabel('TWh/Year Consumption in Niue')
plot.legend(labels=['Niue'])
plot.title("Valores Referentes ao Consumo de Energia Elétrica no Niue")
plot.show()
```

#Valores referentes ao Consumo de Energia Elétrica no EUA 
(País com maior consumo de energia elétrica dentro do estudo disponível)

```
plot.figure(figsize=(12,6))
plot.plot(dados['United States of America'],c='r')
plot.xlabel('Year')
plot.ylabel('TWh/Year Consumption in USA')
plot.legend(labels=['EUA'])
plot.title("Valores Referentes ao Consumo de Energia Elétrica no EUA")
plot.show()
```

```
plot.figure(figsize=(12,6))
plot.plot(dados['United States of America'],c='g', label = 'USA')
plot.plot(dados['China'],c='r', label = 'China')
plot.xlabel('Year')
plot.ylabel('TWh/Year Consumption')
plot.title("Valores Referentes ao Consumo de Energia Elétrica na China X EUA")
plot.legend()
plot.show()
```

# **INVESTIMENTOS**

O investimento global em energia se recuperou em quase 10% em 2021 e
alcançando US$ 1,9 trilhão. Representando uma reversão da maior parte da queda do ano 2020 causada pela pandemia de covid-19. Apesar disso, os gastos com transições de energia limpa precisam acelerar muito mais rapidamente para atender as metas climáticas, de acordo com a um novo relatório da Agência Internacional de Energia (AIE).

A estimativa é de que o investimento no setor de energia elétrica global teve aumento de cerca 5% no ano 2021 para mais de US$ 820 bilhões, seu nível mais alto, depois de permanecer estável em 2020. As energias renováveis ​​estão dominando o investimento em nova capacidade de geração e foram representados em 70% do total neste ano.
